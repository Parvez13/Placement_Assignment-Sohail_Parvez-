```python
import pandas as pd 
```

```python

df = pd.read_excel("question3_data.xlsx")
```

```python
# Get all Pokemons whose spawn rate is less than 5%(0.05)
df[df['spawn_chance'] < 0.05]
```

```python
# Get all Pokemons that have less than 4 weaknesses
weakness = df['weaknesses'].tolist()
```

```python
# Here we can see that list contains strings, inside a string we have list values.
weakness
```

```python
import ast
# Now we want only list not strings.
weakness_cleaned = [ast.literal_eval(item) for item in weakness]
```

```python
weakness_cleaned 
```

```python
df['weakness_cleaned'] = weakness_cleaned 
```

```python
df.drop('weaknesses',axis=1,inplace=True)
```

```python
df['weakness_cleaned']
```

```python
# Filter out pokemon whose weakness is less than 4
df[df['weakness_cleaned'].apply(lambda x: len(x) < 4)]
```

```python
# Fill NaN value 
df['multipliers'].fillna(0,inplace=True)
```

```python
#Get all Pokemons that have no multipliers at all
df[df['multipliers']==0]
```

```python
## Fill NaN value with '0'
df['prev_evolution'].fillna(0,inplace=True)
df['next_evolution'].fillna(0,inplace=True)
```

```python
#  Get all Pokemons that do not have more than 2 evolutions
evolution = df[(df['prev_evolution'].apply(lambda x: x!=0)) &
                (df['next_evolution'].apply(lambda x: x!=0))]
```
